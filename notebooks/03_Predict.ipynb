Imprimir la predicción de archivos

```python
import pandas as pd
from tabulate import tabulate
import requests
```

```python
url_t = "https://worldcup26.ir/get/teams"
response = requests.get(url_t)
teams = response.json()
teams_flag = {team['fifa_code']: team['flag'] for team in teams['teams']}
```

```python
check = pd.read_parquet('../database/ranking_fifa.parquet')
#check[check['Codigo_Pais'].isin(['COD','UZB'])]
```

```python
df = pd.read_parquet('../database/predictions_r2.parquet')
df['Flag'] = df['Team'].map(teams_flag)
df['Estimated_goals'] = df['Goles_Estimados'].copy()
df.sort_values(by=['Date','Venue'], inplace=True)
df['Estimated_goals2'] = df['Model_predict'].astype(int)
df_test = df[['Date','Team','Flag', 'Opponent', 'Model_predict','Estimated_goals', 'Estimated_goals2']]
df_test
```

```python
filtro = df_test['Date'].dt.date == pd.Timestamp('2026-07-19').date()
df_hoy = df_test[filtro][['Team', 'Flag', 'Opponent', 'Estimated_goals','Estimated_goals2','Model_predict']]
```

```python
from IPython.display import HTML

# Definimos una función para convertir URLs en etiquetas img
def path_to_image_html(path):
    return f'<img src="{path}" width="40" >'

# Aplicamos el estilo al DataFrame
# Esto creará una representación HTML donde los links serán imágenes visibles
html_output = df_hoy.to_html(
    escape=False, 
    formatters=dict(Flag=path_to_image_html)
)

# Mostramos el resultado
display(HTML(html_output))
```
